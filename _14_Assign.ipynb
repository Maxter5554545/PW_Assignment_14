```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

logging.info("3 type of Exception errors are presented in Arithmetic class:\n1.OverflowError\n2.ZeroDivisionError\n3.FloatingPointError")
logging.info("ZeroDivisionError is executed when we divide any number by zero & FloatingPointError occurs when floating/decimal values are unable to calculate each other")
try:
    # Example 1: ZeroDivisionError
    result = 2 / 0  # Attempting to divide by zero
except ZeroDivisionError as z:
    logging.error(f"ZeroDivisionError: {z}")

try:
    # Example 3: FloatingPointError
    result = 1.0/0.0  # Intentional floating-point operation error
except FloatingPointError as f:
    logging.error(f)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
logging.info("The LookupError class is used to check that any key or index is presented in the given value or not.")
logging.info("KeyError is used to check whether the required key is presented in the pointer or not if not then the KeyError exception is executed./nIndexError is used to check whether the required Index in the list is presented or not ")
try:
    k = {"ramesh": "from rajasthan", "mrinal": "from Varanasi"}
    print(k["suresh"])
except KeyError as k:
    print(f"The given key '{k}' is not presented to the pointer/dictionary.")
try:
    l = [1, 2, 3]
    print(l[5])
except IndexError as l:
    logging.error(l)
```
